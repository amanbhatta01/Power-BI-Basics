#Power BI Basics

Question 1 : What is Power BI and why is it used in businesses?

 - Power BI is a Business Intelligence (BI) and Data Visualization tool developed by Microsoft. It helps users collect data from different sources, transform it into meaningful information, and create interactive reports and dashboards for better decision-making.

 - why is it used in businesses:-

   Connect to Data:-

   Power BI connects to multiple data sources such as Excel, SQL Server, CSV files, Azure, and online services.

   Clean and Transform Data:-

   Using Power Query, users can remove errors, fill missing values, split columns, merge tables, and perform other data cleaning tasks.

   Analyze Data:-

   Power BI allows users to create relationships between tables and perform calculations using DAX (Data Analysis Expressions).

   Create Reports:-

   Users can build interactive reports using charts, graphs, maps, tables, KPIs, and slicers.

   Share Insights:-

   Reports and dashboards can be published to the Power BI Service and shared with team members for collaboration.

Question 2 : Name and explain the three main components of Power BI.

  - The three main components of Power BI are:-

     1. Power BI Desktop

     2. Power BI Service

     3. Power BI Mobile

    Explanation:-

     1. Power BI Desktop:-

     Power BI Desktop is a free Windows application used to create reports and dashboards. It allows users to connect to various data sources, clean and transform data using Power Query, create relationships between tables, perform calculations using DAX, and design interactive visualizations. After creating reports, they can be published to the Power BI Service.

     Purpose:-

     Connect to data sources.

    Clean and transform data.

    Build data models.

    Create interactive reports and dashboards.

    2. Power BI Service

    Power BI Service is the cloud-based (online) platform where reports and dashboards created in Power BI Desktop are published. It allows users to share reports with others, schedule automatic data refreshes, collaborate with team members, and access reports securely from anywhere through a web browser.

    Purpose:-

    Publish reports online.

    Share dashboards with users.
    
    Schedule automatic data refresh.

    Collaborate with teams.

    Manage workspaces and security.

    3. Power BI Mobile

    Power BI Mobile is the mobile application available for Android and iOS devices. It enables users to access dashboards and reports anytime, receive notifications, and monitor business performance while on the go.

    Purpose:=

    View reports on smartphones and tablets.

    Monitor business performance anywhere.

    Receive alerts and notifications.

    Access real-time dashboards.

Question 3 : Explain the Power BI workflow.

 - The Power BI workflow is the process of collecting data, preparing it, analyzing it, creating reports, and sharing insights with users. It helps organizations convert raw data into meaningful visualizations for better decision-making.

 - The explanation:-

 1. Connect to Data Sources

    The first step is to connect Power BI to different data sources such as Excel files, SQL Server, CSV files, Azure, SharePoint, or online services.

 2. Transform and Clean Data (Power Query)

    After importing the data, Power Query is used to clean and transform it. This includes removing duplicates, handling missing values, changing data types, splitting columns, merging tables, and filtering unwanted data.

 3. Create the Data Model

    Once the data is cleaned, relationships are created between different tables. Measures and calculated columns can also be created using DAX (Data Analysis Expressions).

 4. Build Reports and Dashboards

    Users create interactive reports using charts, tables, maps, cards, slicers, and other visualizations. These visuals help users understand trends and business performance.

 5. Publish to Power BI Service

    The completed report is published from Power BI Desktop to the Power BI Service, making it available online.

 6. Share and Collaborate

    Reports and dashboards are shared with team members, managers, or stakeholders. Permissions can be managed to ensure secure access.

 7. Refresh and Monitor Data

    Power BI supports scheduled or real-time data refresh, ensuring reports always display the latest information.

Question 4 : List any four data cleaning tasks that can be performed in Power Query.

 - Power Query is a data transformation tool in Power BI used to clean and prepare data before analysis. Some common data cleaning tasks are:-

  1. Remove Duplicates

    Power Query can identify and remove duplicate rows from a dataset to ensure each record is unique and improve data accuracy.

    Eliminates repeated records from the dataset.

  2. Handle Missing Values

    Missing or null values can be replaced, filled, or removed using Power Query to make the dataset complete and suitable for analysis.

    Replaces, fills, or removes null values.

  3. Change Data Types

    Power Query allows users to convert data into the correct format, such as changing text to numbers, dates, or currency.

    Converts data into the correct format (Text, Number, Date, etc.).

  4. Split or Merge Columns

    Columns can be split into multiple columns (e.g., Full Name into First Name and Last Name) or merged into a single column when needed.

    Restructures data by separating or combining columns.

Question 5 : Write step by step instructions to load the above dataset.

 - To load the **"Real_Estate_Data_100_Rows.csv"** dataset into Power BI, follow these steps:-

 1. Open Power BI Desktop.

 2. Click on Home tab → Get Data.

 3. From the list of data sources, select Text/CSV.

 4. Click Connect.

 5. Browse to the location where the Real_Estate.csv file is saved.

 6. Select the Real_Estate.csv file and click Open.

 7. A preview window will appear showing the dataset. Verify that the data is displayed correctly.

 8. If the data needs cleaning or transformation, click Transform Data to open Power Query Editor.

 9. Perform any required data cleaning tasks, such as:-

   - Remove duplicate rows.

   - Handle missing values.

   - Change data types.

   - Split the Owner Name column into First Name and Last Name

 10. After completing the transformations, click Close & Apply.

 11. The Real_Estate.csv dataset is now loaded into Power BI and is ready for creating reports and visualizations.

Question 6 : Define Data View, Report View, and Model View. Explain the purpose of each view.

 - Power BI Desktop provides three main views that help users manage data, create reports, and build relationships between tables. These views are Data View, Report View, and Model View.

 1. Data View

    Data View displays the data loaded into Power BI in a tabular format. It allows users to examine records, create calculated columns, and verify the imported data.

    Purpose:-

    View the imported dataset.

    Check data accuracy.

    Create calculated columns.

    Inspect individual records.

 2. Report View

    Report View is the main workspace where users create interactive reports and dashboards using charts, tables, maps, slicers, and other visualizations.

    Purpose:-

    Design interactive reports.

    Create charts and graphs.

    Add filters and slicers.

    Analyze and present business data.

 3. Model View

    Model View shows the relationships between different tables in the dataset. It helps users create and manage the data model.

    Purpose:-

    Create relationships between tables.

    Manage data models.

    View primary and foreign key relationships.

    Improve data analysis performance.

Question 7 : Discuss the different data sources that Power BI supports.

 - Power BI supports a wide variety of data sources, allowing users to collect data from files, databases, cloud services, online platforms, and web sources. This flexibility enables organizations to combine data from multiple systems into a single report for analysis.

 - Different data sources that Power BI supports:-

  1. File-Based Data Sources

    Power BI can import data stored in different file formats.

    Examples:-

    Excel Workbook (.xlsx), CSV (Comma-Separated Values), Text Files (.txt), XML, JSON, PDF.

  2. Database Data Sources

    Power BI can connect directly to relational and enterprise databases.

    Examples:-

    SQL Server, MySQL, Oracle Database, PostgreSQL, Microsoft Access.

  3. Cloud-Based Data Sources

    Power BI connects to cloud storage and cloud databases.

    Examples:-

    Microsoft Azure SQL Database, Azure Blob Storage, Google BigQuery, Snowflake.

  4. Online Services

    Power BI supports many online business applications.

    Examples:-

    SharePoint Online, Dynamics 365, Salesforce, Google Analytics.

  5. Web and API Data Sources

    Power BI can retrieve data from websites and APIs.

    Examples:-

    Web Pages, REST APIs, OData Feeds.

  6. Other Data Sources

    Power BI also supports several other connection types.

    Examples:-

    Folder (multiple files), Power BI Datasets, Blank Query, R Scripts, Python Scripts.
